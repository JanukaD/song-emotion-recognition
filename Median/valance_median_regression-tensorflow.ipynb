```python
X = dataset[['Method of Moments Overall Average Median', 
             'Peak Based Spectral Smoothness Overall Standard Deviation',
             'Relative Difference Function Overall Average',
             'Spectral Rolloff Point Overall Standard Deviation',
             'Strongest Frequency Via Zero Crossings Overall Standard Deviation',
              'Spectral Centroid Overall Standard Deviation',
              'Magnitude Spectrum Overall Standard Deviation Median',
              'Method of Moments Overall Standard Deviation Median',
                           'Strongest Frequency Via Zero Crossings Overall Average',
                           'Zero Crossings Overall Average',
                           'Area Method of Moments of MFCCs Overall Average Median',
                           'Magnitude Spectrum Overall Average Median',
                           'Area Method of Moments Overall Average Median',
                           'MFCC Overall Average Median',
                           'LPC Overall Average Median',
                           'Spectral Rolloff Point Overall Average',
                           'Fraction Of Low Energy Windows Overall Standard Deviation',
                           'Peak Based Spectral Smoothness Overall Average',
                           'Compactness Overall Average',
                           'Compactness Overall Standard Deviation',
                           'Spectral Centroid Overall Average',
                           'Strongest Frequency Via Spectral Centroid Overall Average',
                           'Strongest Frequency Via FFT Maximum Overall Standard Deviation',
                           'Strength Of Strongest Beat Overall Standard Deviation',
                           'Strongest Beat Overall Average',
                           'Spectral Flux Overall Standard Deviation',
                           'Spectral Flux Overall Average',
                           'Power Spectrum Overall Standard Deviation Median',
                           'Power Spectrum Overall Average Median',
                           'Strength Of Strongest Beat Overall Average',
                           'MFCC Overall Standard Deviation Median',
                           'Relative Difference Function Overall Standard Deviation',
                  'Root Mean Square Overall Average']]


y = dataset.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model

input_layer = Input(shape=(X.shape[1],))
dense_layer_1 = Dense(100, activation='relu')(input_layer)
dense_layer_2 = Dense(50, activation='relu')(dense_layer_1)
dense_layer_3 = Dense(25, activation='relu')(dense_layer_2)
output = Dense(1)(dense_layer_3)

model = Model(inputs=input_layer, outputs=output)
model.compile(loss="mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])
```

```python
history = model.fit(X_train, y_train, batch_size=2, epochs=100, verbose=1, validation_split=0.2)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

pred_train = model.predict(X_train)
print(np.sqrt(mean_squared_error(y_train,pred_train)))

pred = model.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred)))
```
